```python
import sklearn
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import load_breast_cancer
# Load dataset
data = load_breast_cancer()
```

```python
print(data)
```

Organize Dataset


```python
data.keys()
```

```python
label_names=data['target_names']
labels=data['target']
feature_names=data['feature_names']
features=data['data']
```

```python
print(label_names)

print(labels)

print(feature_names)

print(features)
```

Organizing Data into Datasets

Splitting the Data

The function for spliting the data randomly splits the data using the test_size
parameter.we now have a test set (test) that represents 33% of the original dataset. 

```python

train, test, train_labels, test_labels= train_test_split(features,labels,test_size=0.33,random_state=42)
```

```python


train, test, train_labels, test_labels= train_test_split(features,labels,test_size=0.33,random_state=42)
```

```python
print(train_test_split)
```

Building and Evaluation of the Model

We will use simple algorithms that usually well in binary classifications such as Naive Bayes(NB)

```python
#initialize the classifier
gnb= GaussianNB()

#training the classifier
model= gnb.fit(train, train_labels)

model
```

Making Predictions using the Trained Data

```python
predictions= gnb.predict(test)

print(predictions)
```

Evaluating the Models Accuracy

```python
accuracy=accuracy_score(test_labels,predictions)
```

```python
# Plotting the accuracy score
plt.bar(['Accuracy'], [accuracy])
plt.ylim(0, 1)  # Set y-axis limit from 0 to 1
plt.ylabel('Score')
plt.title('Model Accuracy')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(['Accuracy'], [accuracy], color='brown')
ax.set_ylim(0, 1)  # Set y-axis limit from 0 to 1
ax.set_ylabel('Score')
ax.set_title('Model Accuracy')
```
